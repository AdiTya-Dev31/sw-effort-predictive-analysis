```python
from scipy.io import arff
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Formatação mais bonita para os notebooks
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (15,5)
```

```python
df_albrecht = pd.read_csv('../Datasets/06.nasa93.csv')
df_albrecht.head()
```

### Cleaning Data

```python
# Convert categorical data to continuous data
df_albrecht.loc[df_albrecht['pcap'] == 'n', 'pcap'] = 1.00
df_albrecht.loc[df_albrecht['pcap'] == 'h', 'pcap'] = 0.86
df_albrecht.loc[df_albrecht['pcap'] == 'vh', 'pcap'] = 0.70

df_albrecht.loc[df_albrecht['time'] == 'n', 'time'] = 1.00
df_albrecht.loc[df_albrecht['time'] == 'h', 'time'] = 1.11
df_albrecht.loc[df_albrecht['time'] == 'vh', 'time'] = 1.30
df_albrecht.loc[df_albrecht['time'] == 'xh', 'time'] = 1.66

df_albrecht.loc[df_albrecht['cplx'] == 'vl', 'cplx'] = 0.70
df_albrecht.loc[df_albrecht['cplx'] == 'l', 'cplx'] = 0.85
df_albrecht.loc[df_albrecht['cplx'] == 'n', 'cplx'] = 1.00
df_albrecht.loc[df_albrecht['cplx'] == 'h', 'cplx'] = 1.15
df_albrecht.loc[df_albrecht['cplx'] == 'vh', 'cplx'] = 1.30
df_albrecht.loc[df_albrecht['cplx'] == 'xh', 'cplx'] = 1.65
```

```python
def print_r2(dependents_var_list, independent_var):
    model = LinearRegression()
    for x in dependents_var_list:
        model.fit(np.reshape(np.array(df_albrecht[x]), (len(df_albrecht[x]), 1)), independent_var) 
        print("R-Squared coefficient of '{}' : {}".format(x,model.score(np.reshape(np.array(df_albrecht[x]), (len(df_albrecht[x]), 1)),independent_var)))


effort = np.reshape(np.array(df_albrecht['act_effort']), (len(df_albrecht['act_effort']), 1)) 
print_r2(['pcap', 'time','cplx'], effort)
```

### Showing Correlation

```python
def show_corr(columnName, xlabel, ylabel):
    model = LinearRegression()
    column = df_albrecht[columnName].values.reshape(-1,1)
    model.fit(column, df_albrecht['act_effort'])
    predictions =  model.predict(column)
    plt.plot(column, predictions, color='red', linewidth=1)
    plt.scatter(column, df_albrecht['act_effort'], alpha=0.4, c=df_albrecht['act_effort'])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
    
show_corr('pcap',"Programmer Capability",'Effort')
show_corr('time',"Time Constraint for CPU",'Effort')
show_corr('cplx',"Process Complexity",'Effort')
```
